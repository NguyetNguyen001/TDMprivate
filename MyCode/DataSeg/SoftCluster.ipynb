```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

```python
# 1. Generate sample data (for illustration)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 2. Fit a Gaussian Mixture Model
gmm = GaussianMixture(n_components=4)  # We assume there are 4 clusters
gmm.fit(X)

# 3. Get the soft clustering (probabilities of membership to each cluster)
probs = gmm.predict_proba(X)

# 4. Get the hard labels (the cluster each data point is most likely assigned to)
labels = gmm.predict(X)

# 5. Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, marker='o', edgecolors='k')
plt.title("Soft Clustering (GMM) - Hard Labels")
plt.show()

# 6. Optional: Visualize the probabilities (for soft clustering)
plt.figure(figsize=(8, 6))
for i in range(4):  # Plot the probability of each point belonging to the ith cluster
    plt.scatter(X[:, 0], X[:, 1], c=probs[:, i], cmap='viridis', s=50, marker='o', alpha=0.5)
    plt.title(f"Probability of Belonging to Cluster {i+1}")
    plt.show()
```
