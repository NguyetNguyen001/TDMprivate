```python
import numpy as np
import matplotlib.pyplot as plt
from skfda import FDataGrid
from skfda.ml.clustering import FuzzyCMeans
from sklearn.metrics import silhouette_score

# Simulate time-series data (replace this with your own dataset)
n_samples = 20  # Number of time series
n_points = 100  # Number of time points per series
t = np.linspace(0, 10, n_points)  # Time grid

# Create synthetic data: 10 sine waves and 10 cosine waves (just for demonstration)
y = np.vstack([np.sin(t + np.random.randn()) for _ in range(n_samples // 2)] + 
              [np.cos(t + np.random.randn()) for _ in range(n_samples // 2)])

# Convert the synthetic data into a functional data object
fd_data = FDataGrid(data_matrix=y, grid_points=t)

# Visualize the data
plt.figure(figsize=(10, 6))
for curve in import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skfda import datasets
from skfda.exploratory.visualization.clustering import (
    ClusterMembershipLinesPlot,
    ClusterMembershipPlot,
    ClusterPlot,
)
from skfda.ml.clustering import FuzzyCMeans, KMeansfd_data:
    plt.plot(t, curve, alpha=0.5, color='gray')
plt.title("Synthetic Time-Series Data (Functional Data)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()

# Create Fuzzy C-Means object (specify the number of clusters)
n_clusters = 2
fcm = FuzzyCMeans(n_clusters=n_clusters, max_iter=100, tol=1e-4)

# Fit the model to the functional data
fcm.fit(fd_data)

# Retrieve the cluster labels
labels = fcm.labels_

# Plot the time series grouped by clusters
for label in np.unique(labels):
    plt.figure(figsize=(10, 6))
    cluster_data = fd_data[labels == label]
    for curve in cluster_data:
        plt.plot(t, curve, alpha=0.5, label=f"Cluster {label + 1}")
    plt.title(f"Cluster {label + 1}")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

# Get the cluster centers (average functions of each cluster)
centers = fcm.centers_

# Plot the cluster centers
plt.figure(figsize=(10, 6))
for i, center in enumerate(centers):
    plt.plot(t, center, label=f"Cluster {i + 1} Center", linewidth=2)
plt.title("Cluster Centers (Average Functions)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()

# Evaluate the clustering quality using the silhouette score
sil_score = silhouette_score(fd_data.data_matrix, labels)
print(f"Silhouette Score: {sil_score}")
```
