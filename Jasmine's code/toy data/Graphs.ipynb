```python
import matplotlib.pyplot as plt
import csv
```

```python
data = []
with open("best_pose_58263e34-a45c-11ee-88ec-eb6a8d5269b4(in).csv",'r') as file:
    csv_file = csv.DictReader(file)
    for row in csv_file:
        data.append(dict(row))
```

```python
numSatsInSolution = [x['numSatsInSolution'] for x in data ]
latitude = [x['latitude'] for x in data]
```

```python
plt.plot(numSatsInSolution,latitude,'co')
plt.show()
```
