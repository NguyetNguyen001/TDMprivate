```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('best_pose_58263e34-a45c-11ee-88ec-eb6a8d5269b4(in).csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5), dpi = 200)
sns.scatterplot(x = 'numSatsInSolution', y = 'longitudeStdDev', data = df)
```

```python
plt.figure(figsize=(10,5), dpi = 200)
sns.scatterplot(x = 'numSatsInSolution', y = 'latitudeStdDev', data = df)
```

```python
plt.figure(figsize=(10,5), dpi = 200)
sns.boxplot(x = 'numSatsInSolution', y = 'latitudeStdDev', data = df)
```

```python
plt.figure(figsize=(10,5), dpi = 200)
sns.scatterplot(x = 'longitudeStdDev', y = 'latitudeStdDev', data = df, hue = 'numSatsInSolution', palette = 'Dark2')
```

```python
plt.figure(figsize=(10,5), dpi = 200)
sns.scatterplot(x = 'latitude', y = 'longitude', data = df, hue = 'numSatsInSolution', palette = 'Dark2')
```
