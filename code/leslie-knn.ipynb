```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Select features and target from chassis_data
features = ['yawRate', 'throttlePercentage', 'steeringPercentage', 'wheelSpeed.wheelSpdFr']
X = chassis_data[features]
y = chassis_data['BinaryDrivingMode']  # Assuming this is your target variable for disengagements

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Creating and training the KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Making predictions
y_pred = knn.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("KNN_Confusion Matrix_Disengagements")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report


# Select relevant features and the target variable
X = gmID_130_chassis_df[['brakePercentage', 'throttlePercentage', 
                         'steeringPercentage', 'speedMps', 
                         'signal.turnSignal']]
y = gmID_130_chassis_df['BinaryDrivingMode']
```

```python
 # One-Hot Encode the categorical variables
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_encoded = encoder.fit_transform(X[['signal.turnSignal']])

# Convert the encoded data to a DataFrame
X_encoded_df = pd.DataFrame(
    X_encoded, 
    columns=encoder.get_feature_names_out(['signal.turnSignal'])
)

# Combine the encoded columns with the continuous variables
X_continuous = X[['brakePercentage', 'throttlePercentage', 
                  'steeringPercentage', 'speedMps']].reset_index(drop=True)

X_final = pd.concat([X_encoded_df, X_continuous], axis=1)

# Standardize the continuous variables to ensure consistent scaling
scaler = StandardScaler()
X_final_scaled = scaler.fit_transform(X_final)

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_final_scaled, y, 
                                                    test_size=0.2, random_state=42)

# Initialize and train the KNN model (choose k=5 as an example)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn.predict(X_test)

# Evaluate the model using accuracy and classification report
print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
# Plot Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
class_labels = ['ManualMode (0)', 'AutoMode (1)']
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)

# Plot the confusion matrix with a title
disp.plot(cmap=plt.cm.Oranges)
plt.title("KNN Confusion Matrix - gmID_130_chassis_df")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd

# Placeholder for the results
results = []

# Loop through each gmID in the dataset
for gm_id in merged_df['groupMetadataID_x'].unique():  # Assuming 'gmID' identifies each run
    print(f"Processing gmID: {gm_id}")
    
    # Filter the data for the current gmID
    gm_data = merged_df[merged_df['groupMetadataID_x'] == gm_id]
    
    # Split features and target
    X = gm_data[['brakePercentage', 'throttlePercentage', 'steeringPercentage', 'speedMps']]
    y = gm_data['BinaryDrivingMode']
    
    # Standardize the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    # Initialize and train the KNN model
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)
    
    # Make predictions
    y_pred = knn.predict(X_test)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    class_report = classification_report(y_test, y_pred, output_dict=True)  # Get detailed report
    
    # Save the results for this gmID
    results.append({
        'gmID': gm_id,
        'accuracy': accuracy,
        'classification_report': class_report
    })

    # Print the results for this gmID
    print(f"Accuracy for gmID {gm_id}: {accuracy:.4f}")
    print(f"Classification Report for gmID {gm_id}:\n{classification_report(y_test, y_pred)}\n")

# Convert results into a DataFrame for analysis
results_df = pd.DataFrame(results)

# Display or save the results
print(results_df)
```
