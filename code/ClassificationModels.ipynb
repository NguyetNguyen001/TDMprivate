```python
train_ids = blue_df["groupMetadataID_x"].unique()[:train_count]
test_ids = blue_df["groupMetadataID_x"].unique()[train_count:]
```

## Decision Tree

```python
from sklearn import tree
X_train = blue_df[ blue_df['groupMetadataID_x'].isin(train_ids)][d_cols]
Y_train = blue_df[ blue_df['groupMetadataID_x'].isin(train_ids)][r_cols]

X_test = blue_df[ blue_df['groupMetadataID_x'].isin(test_ids)][d_cols]
Y_test = blue_df[ blue_df['groupMetadataID_x'].isin(test_ids)][r_cols]


clf = tree.DecisionTreeClassifier(max_depth=5)
clf = clf.fit(X_train, Y_train)
```

```python
prediction = clf.predict(X_test)
prediction
#blue_df[d_cols].iloc[0]
```

```python
results= pd.DataFrame()
results["predicted"] = prediction
results['actual'] = list(Y_test['TernaryDrivingModeTransition'])
results
```

```python
from sklearn.metrics import confusion_matrix

tn, fp, fn, tp = confusion_matrix(Y_test, prediction).ravel()

p = tp + fn
n = tn + fp


tpr = tp/p
tnr = tn/p

acccuracy = (tp+tn)/(tp+fp+tn+fn)
balanced_acccuracy = (tpr+tnr)/2


print(f"p: {p}")
print(f"n: {n}")
print(f"tp: {tp}")
print(f"fp: {fp}")
print(f"tn: {tn}")
print(f"fn: {fn}")

print(f"tpr: {tpr}")
print(f"tnr: {tnr}")


print(f'acccuracy: {acccuracy}')
print(f'balanced_acccuracy: {balanced_acccuracy}')
```
