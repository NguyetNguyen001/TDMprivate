## Decision Tree


### 1

Purely chassis


```python
from sklearn import tree
X_train = blue_df[blue_df['groupMetadataID_x'].isin(train_ids)][d_cols]
Y_train = blue_df[blue_df['groupMetadataID_x'].isin(train_ids)][r_cols]

X_test = blue_df[blue_df['groupMetadataID_x'].isin(test_ids)][d_cols]
Y_test = blue_df[blue_df['groupMetadataID_x'].isin(test_ids)][r_cols]


clf = tree.DecisionTreeClassifier(max_depth=5)
clf = clf.fit(X_train, Y_train)
```

```python
prediction = clf.predict(X_test)
prediction
# blue_df[d_cols].iloc[0]
```

```python
from sklearn.metrics import confusion_matrix, precision_recall_curve, accuracy_score, f1_score, balanced_accuracy_score

tn, fp, fn, tp = confusion_matrix(Y_test, prediction).ravel()

# p = tp + fn
# n = tn + fp
#
#
# tpr = tp / p
# tnr = tn / p
# print(f"p: {p}")
# print(f"n: {n}")
# print(f"tp: {tp}")
# print(f"fp: {fp}")
# print(f"tn: {tn}")
# print(f"fn: {fn}")
#
# print(f"tpr: {tpr}")
# print(f"tnr: {tnr}")

acccuracy = accuracy_score(Y_test, prediction)
balanced_acccuracy = balanced_accuracy_score(Y_test, prediction)
f1 = f1_score(Y_test, prediction)
precision, recall, thresholds = precision_recall_curve(Y_test, prediction)


print(f"precision: {precision[1]}")
print(f"recall: {recall[1]}")
print(f"f1: {f1}")

print(f'acccuracy: {acccuracy}')
print(f'balanced_acccuracy: {balanced_acccuracy}')
```
