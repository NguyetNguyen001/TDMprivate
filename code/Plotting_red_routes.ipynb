```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
```

### Read and preprocces data

```python
gmID = '6d2ea45a-c839-11ee-a7fc-dd032dba19e8'


df_chassis = pd.read_csv(f'./data/Raw_Data/{gmID}/_apollo_canbus_chassis/{gmID}_apollo_canbus_chassis.csv')
df_pose = pd.read_csv(f'./data/Raw_Data/{gmID}/_apollo_sensor_gnss_best_pose/{gmID}_apollo_sensor_gnss_best_pose.csv')
drive_event = pd.read_csv(f'./data/drive_event/drive_events.csv')

#df_metadata = pd.read_csv(f'./data/metadata/{gmID}.csv')
ads_index = pd.read_csv('./data/ads_data_index.csv')
```

```python
ads_index = ads_index[ads_index['groupMetadataID'] == gmID]

drive_event = drive_event[drive_event['groupMetadataID'] == gmID]

drive_event.sort_values(by='time', inplace=True)
df_chassis.sort_values(by='time', inplace=True)
df_pose.sort_values(by='time', inplace=True)

df_chassis['drivingMode'] = np.where(df_chassis['drivingMode'] == 'COMPLETE_AUTO_DRIVE', 'COMPLETE_AUTO_DRIVE','COMPLETE_MANUAL')
```

### Map route

```python
ll = df_pose.copy()
ll = ll.sort_values('time')
ll['time'] = ll['time'].astype(float)
ll['time'] = ll['time'].astype('datetime64[ns]')
# ll = ll[["groupMetadataID", 'time', 'latitude', 'longitude']]
# ll = ll[["groupMetadataID", 'time', 'latitude', 'longitude', 'longitudeStdDev', 'latitudeStdDev','numSatsTracked', 'numSatsInSolution', 'numSatsMulti','solStatus']]


merged_df = pd.merge(df_chassis, ll, on='time')
```

```python
# Define your map with your dataframe name first and keep everything else the same
fig = px.scatter_mapbox(merged_df, lat="latitude",
                        lon="longitude",
                        hover_data=["time", "drivingMode", "transitionType"],
                        color="drivingMode",
                        zoom=13,
                        height=1000,
                        width=1200,
                        # color_discrete_map={
                        #   "M->A": "#63ACBE",
                        #   "A->M": "#601A4A",
                        #   "ERROR": "#EE442F",
                        #   "NONE":"yellowgreen",
                        # }
                        color_discrete_map={
                            "COMPLETE_MANUAL": "#601A4A",
                            "COMPLETE_AUTO_DRIVE": "#63ACBE",
                            "EMERGENCY_MODE": "#EE442F",
                        }

                        )

# Layout settings
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})

fig.update_traces(marker=dict(size=8))

# Display the map
fig.show()
```
