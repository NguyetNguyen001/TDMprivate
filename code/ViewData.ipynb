# BOTO & PandasGUI

### **Goal:** Access dataset and metadata from DynamoDB. View both with PandasGUI


```python
import boto3
import pandas as pd

from pandasgui import show

from boto3.dynamodb.conditions import Key, Attr
```

Connecting to the dataset


```python
client = boto3.client('dynamodb')

dynamodb = boto3.resource('dynamodb')

mainTable = dynamodb.Table('ads_passenger_processed')
metaTable = dynamodb.Table('ads_passenger_processed_metadata')
```

Pull sample data using the scan command


```python
mainResponse = mainTable.scan(FilterExpression=Attr('time').gt(0))
metaResponse = metaTable.scan(FilterExpression=Attr('time').gt(0))

print(dict.keys(mainResponse))
print(dict.keys(metaResponse))
```

Coerce datasets into a pandas DataFrame


```python
main_df = pd.DataFrame.from_dict(pd.json_normalize(
    mainResponse['Items']), orient='columns')
meta_df = pd.DataFrame.from_dict(pd.json_normalize(
    metaResponse['Items']), orient='columns')
```

```python
main_df.head()
```

View dataframes with PandasGUI


```python
show(main_df)
show(meta_df)
```
