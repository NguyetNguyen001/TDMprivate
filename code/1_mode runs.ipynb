```python
import CalculatedFieldSubroutines as cfs
import pandas as pd
import numpy as np
import os
import plotly.express as px
```

### Read and preprocces data


```python
ads_index = pd.read_csv('./data/ads_data_index.csv')
```

```python
d = []

for gmID in ads_index["groupMetadataID"]:
    df_chassis = pd.read_csv(f'./data/Raw_Data/{gmID}/_apollo_canbus_chassis/{gmID}_apollo_canbus_chassis.csv')
    cfs.BinaryDrivingMode(df_chassis)
    print({"id": gmID, "unique":df_chassis["BinaryDrivingMode"].unique()})
    
    d.append({"id": gmID, "unique":df_chassis["BinaryDrivingMode"].unique()})
    
```

```python
p = pd.DataFrame(d)
p
```

```python
p["isOneMode"] = p["unique"].apply(len) == 1

p
```

```python
ads_index["isOneMode"] = np.where(ads_index["groupMetadataID"].isin(p[p["isOneMode"] == True]["id"]),True,False)

ads_index[["groupMetadataID","isOneMode"]]
```

```python
ads_index.to_csv('./data/ads_data_index.csv', index=False)
```
