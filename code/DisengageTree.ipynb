```python
red_dfs = []

for gmID in gmIDs:

    if ( cfs.give_route( gmID ) != 'Red' ):

        continue

    df = cfs.retrieve_gmID_topic( gmID, '/apollo/canbus/chassis' )

    cfs.BinaryDrivingMode( df )

    red_dfs.append( df )
```

```python
green_dfs = []

for gmID in gmIDs:

    if ( cfs.give_route( gmID ) != 'green' ):

        continue

    df = cfs.retrieve_gmID_topic( gmID, '/apollo/canbus/chassis' )

    cfs.BinaryDrivingMode( df )

    green_dfs.append( df )
```
